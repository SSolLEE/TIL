## Time_Series
### 주가 데이터

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv")
df.info()
```

- Yr_Mo_Dy를 판다스에서 인식할 수 있는 datetime64 타입으로 변경하기

```python
df.Yr_Mo_Dy = pd.to_datetime(df.Yr_Mo_Dy)
df.info()
```

- Yr_Mo_Dy에 존재하는 연도의 유일값을 모두 출력하기

```python
df.Yr_Mo_Dy.dt.year.unique()
```

- Yr_Mo_Dy에 연도가 2061년 이상의 경우에는 모두 잘못된 데이터이므로, 해당 경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하기

```python
def fix_year(x):
    import datetime

    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_year)
df.head()
```

- 연도별 각 컬럼의 평균값 구하기

```python
df.groupby(df.Yr_Mo_Dy.dt.year).mean()
```

- weekday 컬럼을 만들고 요일별로 매핑하기
(월 : 0, 화 : 1 - 일 : 6)

```python
df['weekday'] = df.Yr_Mo_Dy.dt.weekday
df['weekday'].to_frame().head(3)
```

- weekday 컬럼을 기준으로 주말이면 1, 평일이면 0의 값을 가지는 WeekCheck 컬럼을 만들기


```python
df['WeekCheck'] = df['weekday'].map(lambda x : 1 if x in [5, 6] else 0)
df['WeekCheck'].to_frame().head(3)
```

- 연도, 일자 상관없이 모든 컬럼의 각 달의 평균 구하기

```python
df.groupby(df.Yr_Mo_Dy.dt.month).mean()
```

- 모든 결측치는 컬럼 기준 직전의 값으로 대체하고 첫 번째 행에 결측치가 있을 경우 뒤에 있는 값으로 대체하기

```python
df = df.fillna(method='ffill').fillna(method='bfill')
df.isnull().sum()
```

- 연도 - 월을 기준으로 모든 컬럼의 평균값을 구하기

```python
df.groupby(df.Yr_Mo_Dy.dt.to_period('M')).mean().head()
```

- RPT 컬럼의 값을 일자별 기준으로 1차 차분하기

```python
df['RPT']
```

```python
df['RPT'].diff()
```

- RPT와 VAL 컬럼을 일주일 간격으로 각각 이동 평균한 값을 구하기

```python
df.RPT, df.VAL
```

### 미세먼지 데이터

```python
data = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv")
data.info()
```

```python
data.head(3)
```

- (년-월-일:시) 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하기. 서울시의 제공 데이터의 경우 0시가 24시로 표현됨

```python
import pandas as pd

def change_date(x):
    if isinstance(x, str):
        hour = x.split(':')[1]
        date = x.split(":")[0]

        if hour == '24':
            hour = '00:00:00'
            final_date = pd.to_datetime(date + " " + hour) + pd.Timedelta(days=1)
        else:
            hour = hour + ':00:00'
            final_date = pd.to_datetime(date + " " + hour)

        return final_date
    else:
        return x

# Assuming 'data' is your DataFrame
data['(년-월-일:시)'] = data['(년-월-일:시)'].apply(change_date)
data.head(2)
```

- 일자별 영어요일 이름을 dayName 컬럼에 저장하기

```python
data['dayName'] = data['(년-월-일:시)'].dt.day_name()
data['dayName']
```

- 일자별 각 PM10등급의 빈도수 파악하기

```python
data1 = data.groupby(['dayName', 'PM10등급'], as_index=False).size()
data1.head()
```

```python
data2 = data1.pivot(index='dayName', columns= 'PM10등급', values = 'size').fillna(0)
data2
```

- 시간이 연속적으로 존재하며 결측치가 없는 지 확인하기

```python
# 시간을 차분했을 경우 첫 값은 Nan, 이후 모든 차분 값이 동일하면 연속이라 판단한다.
check = len(data['(년-월-일:시)'].diff().unique())
if check == 2:
    data = True
else:
    data = False

data
```
