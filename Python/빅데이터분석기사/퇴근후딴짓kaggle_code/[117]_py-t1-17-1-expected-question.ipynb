```python
# 라이브러리 불러오기
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터 불러오기
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# datetime으로 type변경
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
# 새로운 컬럼 추가 (년, 월, 일)
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df.head()
```

```python
# 문제 조건에 맞는 값 구하기
ans = df[(df['Year'] == 2022) & (df['Month'] == 5)]
med = ans['Sales'].median()
med
```
