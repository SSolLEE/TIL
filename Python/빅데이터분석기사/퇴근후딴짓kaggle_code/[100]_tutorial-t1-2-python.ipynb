## 문제1
- 데이터셋(basic1.csv)의 'f5' 컬럼을 기준으로 상위 10개의 데이터를 구하고,
- 'f5'컬럼 10개 중 최소값으로 데이터를 대체한 후, 
- 'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
df.head()
```

```python
# f5컬럼을 기준으로 내림차순 정렬
A = df.sort_values(by = 'f5', ascending=False).iloc[:10]
A
```

```python
# 최소값 찾기
min = A['f5'].min()
min
```

```python
# 최소값으로 데이터 대체하기
df.iloc[:10, -1] = min
df.head(10)
```

```python
# 80세 이상의 f5컬럼 평균
age_80 = df[df['age'] >= 80].f5.mean()
age_80
```

## 문제2
- 데이터셋(basic1.csv)의 앞에서 순서대로 70% 데이터만 활용해서,
- 'f1'컬럼 결측치를 중앙값으로 채우기 전후의 표준편차를 구하고
- 두 표준편차 차이 계산하기

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
```

```python
# df.info() # row == 100
data70 = df.iloc[:70]
data70
```

```python
## 결측치 확인
data70.isnull().sum()
```

```python
## 결측치 채우기 전 f1컬럼 표준편자
std1 = data70['f1'].std()
```

```python
## 중앙값 확인
med = data70['f1'].median()
```

```python
## 중앙값으로 채우기
data70['f1'] = data70['f1'].fillna(med)

## 다른 방법들
# data70['f1']= data70['f1'].replace(np.nan, med)
# data70 = data70.fillna(value=med)
```

```python
## 결측치 확인
data70.isnull().sum()
```

```python
## 결측치를 채운 후 표준편차 구하기
std2 = data70['f1'].std()
std2
```

```python
print(std1-std2)
```

## 문제3
- 데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오!
- 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
```

```python
df.head(2)
```

```python
mean = df['age'].mean()
std = df['age'].std()
min = mean - 1.5 * std
max = mean + 1.5 * std
# 이상치 age합
sum = df[(df['age'] < min ) | (df['age'] > max)]['age'].sum()
sum

# 다르게 작성방법
# df.loc[(df['age'] > max)]['age'].sum() + df.loc[(df['age']< min)]['age'].sum()
```
