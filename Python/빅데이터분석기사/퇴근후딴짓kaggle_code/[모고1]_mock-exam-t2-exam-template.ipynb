```python
X_train = X_train.drop('enrollee_id', axis=1)
id = X_test.pop('enrollee_id')
```

```python
X_train = X_train.fillna("X")
X_test = X_test.fillna("X")
```

```python
X_train.shape, X_test.shape
```

```python
del_col = ['experience', 'last_new_job', 'company_size']
X_train = X_train.drop(columns = del_col)
X_test = X_test.drop(columns = del_col)
```

```python
from sklearn.preprocessing import RobustScaler, LabelEncoder
rb = RobustScaler()
cols = X_train.select_dtypes(exclude='object').columns
X_train[cols] = rb.fit_transform(X_train[cols])
X_test[cols] = rb.transform(X_test[cols])

cols = X_train.select_dtypes(include='object').columns
for col in cols:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col] = le.transform(X_test[col])
# X_train[cols] = pd.get_dummies(X_train[cols])
# X_test[cols] = pd.get_dummies(X_test[cols])
```

```python
y = y_train['target'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y, test_size=0.2, random_state=11)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 12, max_depth=7, random_state=11)
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
pred_proba = rf.predict_proba(X_val)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_val, pred)
```
