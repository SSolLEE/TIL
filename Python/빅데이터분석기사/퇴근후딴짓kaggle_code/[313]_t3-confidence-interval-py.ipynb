## 점 추정과 구간 추정
데이터셋은 어떤 도시의 일일 평균 온도 입니다.

- 점추정: 데이터셋을 기반으로 이 도시의 평균 연간 온도를 점추정하세요. (반올림하여 소수 둘째자리까지)
- 구간추정: 95% 신뢰수준에서 이 도시의 평균 연간 온도에 대한 신뢰구간을 구하세요.  (반올림하여 소수 둘째자리까지)

```python
import pandas as pd
temperature_data = pd.read_csv("/kaggle/input/bigdatacertificationkr/daily_temperatures.csv")
```

```python
temperature_data.head(2)
```

```python
# 점 추정
mean = temperature_data['Daily Average Temperature'].mean()
print(round(mean,2))
```

```python
# 구간 추정
from scipy import stats

confidence_level = 0.95
sample_std = temperature_data['Daily Average Temperature'].std(ddof=1) # 자유도 n-1
n_samples = len(temperature_data)

# 95% 신뢰구간
interval = stats.t.interval(confidence_level, df=n_samples-1, loc = mean, scale=sample_std/(n_samples**0.5))
print(round(mean, 2))
interval
```

```python
import pandas as pd
from scipy import stats
# CSV 파일 불러오기
temperature_data = pd.read_csv("/kaggle/input/bigdatacertificationkr/daily_temperatures.csv")

# 점추정: 샘플 평균 계산
sample_mean = temperature_data['Daily Average Temperature'].mean()

# 구간추정: 샘플 표준편차 계산 및 신뢰구간 계산
confidence_level = 0.95
sample_std = temperature_data['Daily Average Temperature'].std(ddof=1)  # 자유도 n-1로 설정
n_samples = len(temperature_data)

# 95% 신뢰구간 계산
confidence_interval = stats.t.interval(confidence_level, df=n_samples-1, loc=sample_mean, scale=sample_std/(n_samples**0.5))

sample_mean, confidence_interval
```
