## 주어진 데이터에서 2022년 5월 주말과 평일의 sales컬럼 평균값 차이를 구하시오 (소수점 둘째자리까지 출력, 반올림)
- 데이터셋 : basic2.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 불러오기 (datatime컬럼 지정)
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv", parse_dates=['Date'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.dayofweek # 월~일 == 0~6
```

```python
df.head()
```

```python
# 2022년 5월 주말 데이터
df_wkend = df[(df['dayofweek'] >= 5) & (df['Year'] == 2022) & (df['Month'] == 5)]
# 2022년 5월 평일 데이터
df_wkday = df[(df['dayofweek'] < 5) & (df['Year'] == 2022) & (df['Month'] == 5)]
```

```python
mean_wkend = df_wkend['Sales'].mean()
mean_wkday = df_wkday['Sales'].mean()
print(round(abs(mean_wkday - mean_wkend), 2))
```
