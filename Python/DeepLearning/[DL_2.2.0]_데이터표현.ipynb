### 스칼라(랭크-0 텐서)
- 스칼라(scalar) : 하나의 숫자만 담고 있는 텐서
- 스칼라 텐서, 랭크-0 텐서, 0D 텐서라고도 부름
- 넘파이에서 float32나 float64 타입의 숫자는 스칼라 텐서 또는 배열 스칼라라고 부름
- ndim - 넘파이 배열의 축 갯수 확인 / 스칼라 텐서의 축 갯수 = 0

```python
import numpy as np
x = np.array(12)
x
```

```python
x.ndim
```

### 벡터(랭크-1 텐서)

```python
x = np.array([12, 3, 4, 2, 5])
x
```

```python
x.ndim
```

### 데이터 다시 보기

```python
from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images.shape, train_images.ndim, train_images.dtype
```

### 이미지 확인하기
- 위 코드에서 확인한 랭크-3 텐서에서 여덟 번째 샘플을 확인해본다.

```python
import matplotlib.pyplot as plt
%matplotlib inline

digit = train_images[7]
plt.imshow(digit, cmap=plt.cm.binary)
plt.show()
```

```python
train_labels[7]
```

- 다른 코드로 같은 결과 만들기

```python
fig, ax = plt.subplots()
ax.imshow(digit, cmap=plt.cm.binary)
plt.show()
```

### 넘파이로 텐서 조작하기
- 다음 세 줄의 코드는 같은 결과를 나타내는 서로 다른 코드이다.

```python
my_slice = train_images[10:100]     # 90개의 이미지를 가져오겠다.
my_slice.shape
```

```python
my_slice = train_images[10:100, :, :]    # 90개의 이미지를 가져오겠다.
my_slice.shape
```

```python
my_slice = train_images[10:100, 0:28, 0:28]    # 90개의 이미지를 가져오겠다.
my_slice.shape
```

- 이미지의 오른쪽 아래 14 * 14 픽셀 선택하기

```python
my_slice = train_images[:, 14:, 14:]
my_slice.shape
```

- 음수 인덱스 사용하기

```python
my_slice = train_images[:, 7:-7, 7:-7]
my_slice.shape
```
