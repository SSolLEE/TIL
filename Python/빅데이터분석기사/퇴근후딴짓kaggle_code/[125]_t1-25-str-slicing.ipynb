## basic1 데이터에서 f4가 E로 시작하면서 부산에 살고 20대인 사람은 몇 명일까요?

```python
import pandas as pd
df = pd.read_csv("../input/bigdatacertificationkr/basic1.csv")
```

```python
df.head(3)
```

```python
con1 = df[df['f4'].str[:1] == 'E']
con2 = df[df['city'] == '부산']
con3 = df[df['age'] >= 20]
con4 = df[df['age'] < 30]
print(len(df[con1 + con2 + con3 + con4]))
```
