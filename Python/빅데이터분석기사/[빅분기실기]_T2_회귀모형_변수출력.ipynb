```python
import pandas as pd
df = pd.read_csv('/content/Titanic.csv')
df.head(2)
```

### Gender, sibsp, parch, fare를 독립변수로 사용하여 회귀모형을 실시하였을 때, parch변수의 계수값은?
(반올림하여 소수 셋째 자리까지 계산)
- 해당 독립변수가 종속변수에 미치는 영향의 크기를 나타냄


```python
from sklearn.linear_model import LinearRegression

# Gender 변수를 원-핫 인코딩
df = pd.get_dummies(df, columns=['Gender'], drop_first = True)

# LinearRegression 모델 생성
model = LinearRegression()

X = df[['Pclass', 'Gender_male', 'Age', 'SibSp', 'Parch', 'Fare']] # 독립 변수 설정
y = df['Survived'] # 종속 변수 설정

# X 결측치 처리
X = X.fillna(X.Age.mean())

# 모델 피팅
model.fit(X, y)

# Parch 변수의 계수 출력
parch_coefficient = model.coef_[X.columns.get_loc('Parch')]
print(f'Parch 변수의 계수: {parch_coefficient:.3f}')
```
