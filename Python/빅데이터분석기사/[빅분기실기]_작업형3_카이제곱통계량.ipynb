```python
import pandas as pd
df = pd.read_csv("/content/Titanic.csv")
df.head()
```

### Gender와 Survived변수 간의 독립성 검정을 실시하였을 때, 카이제곱 통계량은? (반올림하여 소수 셋째 자리까지 계산)

- 카이제곱 검정 : 두 범주형 변수가 종속 사건인지, 독립 사건인지 판별하는 것

```python
from scipy.stats import chi2_contingency
# 교차테이블 생성
cross_table = pd.crosstab(df.Gender, df.Survived)
cross_table
```

```python
result = chi2_contingency(cross_table)
result
```

```python
print(f"카이제곱 통계량: {result[0]:.2f}")
```
