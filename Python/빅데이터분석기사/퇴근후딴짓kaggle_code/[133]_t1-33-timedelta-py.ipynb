    - 새벽: 0시 부터 6시 이전
    - 오전: 6시 부터 12시 이전

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/website.csv")
```

# 1. 세션의 지속기간 중 가장 긴 것 반올림하여 분만 출력하기

```python
df['StartTime'] = pd.to_datetime(df['StartTime'])
df['EndTime'] = pd.to_datetime(df['EndTime'])
```

```python
df['diff'] = df['EndTime'] - df['StartTime']
df.head()
```

```python
df['diff'] = pd.to_timedelta(df['diff'])
df['diff'] = df['diff'].dt.total_seconds()/60
df.head()
```

```python
print(int(round(df['diff'].max(), 0)))
```

# 2. 가장 많이 머무른 page, 머문 시간의 평균

```python
df['Page'].value_counts() # Page5
mean = df[df['Page'] == 'Page5']['diff'].mean()
print(int(round(mean/60, 0)))
```

# 3. 사용자들이 가장 활발히 활동하는 시간대(예: 새벽, 오전, 오후, 저녁)를 분석하세요. 이를 위해 하루를 4개의 시간대로 나누고 각 시간대별로 가장 많이 시작된 세션의 수를 계산하고, 그 중에 가장 많은 세션 수를 출력하시오
    - 새벽: 0시 부터 6시 이전
    - 오전: 6시 부터 12시 이전
    - 오후: 12 부터 18시 이전
    - 저녁: 18시 부터 0시 이전

```python
bins = [0, 360, 720, 1080, 1440]
labels = ['새벽', '오전', '오후', '밤']
df['act'] = pd.cut(df['StartTime'].dt.hour * 60 + df['StartTime'].dt.minute, bins = bins, labels = labels)
df.head(2)
```

```python
df['act'].value_counts().max()
```

# 4. user가 가장 많이 접속한 시간

```python
df['StartTime'] = df['StartTime'].astype('str')
df['often'] = df['StartTime'].str.split(" ").str[0]
df.head(2)
```

```python
result = df['often'].value_counts().index[0]
print(result)
```
