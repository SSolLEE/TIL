## 문제: 두 교육 방법의 효과 비교
연구자는 두 가지 다른 교육 방법이 학생들의 성적에 미치는 영향을 비교하고자 합니다. 연구자는 무작위로 선발된 20명의 학생들을 두 그룹으로 나누어 한 그룹에는 교육 방법 A를, 다른 그룹에는 교육 방법 B를 적용합니다. 교육이 끝난 후, 두 그룹의 성적을 비교하기 위해 독립 표본 t-검정과 ANOVA F-검정을 실시하려고 합니다.

다음은 두 그룹의 성적입니다:
다음의 두 가지 검정을 사용하여 두 교육 방법 간의 성적 차이가 통계적으로 유의한지를 검증하세요

1. 독립 표본 t-검정을 실시하여 t 통계량을 구하세요.
2. 독립 표본 t-검정을 실시하여 p-값을 구하세요.
3. ANOVA F-검정을 실시하여 F 통계량을 구하세요.
4. ANOVA F-검정을 실시하여 p-값을 구하세요.


```python
import pandas as pd
df = pd.DataFrame({
    'A':[77, 75, 82, 80, 81, 83, 84, 76, 75, 87],
    'B':[80, 74, 77, 79, 71, 74, 78, 69, 70, 72],
})
df
```

### 독립 표본 t-검정

```python
import scipy.stats as stats

# 독립 표본 t-검정 수행 : ttest_ind 
t_statistic, p_value = stats.ttest_ind(df['A'], df['B'])

# 결과 출력
print('t-통계량 : ', t_statistic)
print('p-value : ', p_value)
```

```python
# p_value를 통해 유의수준에 대한 가설 검정을 수행
alpha = 0.05 # 유의수준
if p_value < alpha:
    print("두 그룹의 성적은 유의하게 다릅니다.")
else:
    print("두 그룹의 성적은 유의한 차이가 없습니다.")
```

### ANOVA F-검정

```python
import scipy.stats as stats

# ANOVA F-검정 수행
f_statistic, p_value = stats.f_oneway(df['A'], df['B'])

print('F-통계량 : ', f_statistic)
print('p-value : ', p_value)
```

```python
# p-value를 통해 유의수준에 대한 가설 검정을 수행
alpha = 0.05  # 유의수준
if p_value < alpha:
    print("그룹 간에 성적 차이는 유의하게 다릅니다.")
else:
    print("그룹 간에 성적 차이는 유의한 차이가 없습니다.")
```

### 해설
