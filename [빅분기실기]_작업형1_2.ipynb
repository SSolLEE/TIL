### GROUPING
- 데이터 불러오기

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv")
df.head(5)
```

- 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하기

```python
df.groupby('host_name').size().head()
```

```python
df.host_name.value_counts().sort_index().head()
```

- 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임 만들기. 빈도수 컬럼의 이름은 counts로 하기

```python
data = df.groupby('host_name').size().\
to_frame().rename(columns={0:'counts'}).\
sort_values('counts', ascending=False)
data.head(2)
```

- neighbourhood_group의 값에 따른 neighbourhood 컬럼의 값의 갯수

```python
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().head()
```

- neighbourhood_group의 값에 따른 neighbourhood 컬럼 값 중 neighbourhood_group의 최댓값 출력

```python
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().\
groupby(['neighbourhood_group'], as_index=False).max()
```

- neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최솟값

```python
df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
```

- neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최솟값 구하기

```python
df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
```

- neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균

```python
df[['neighbourhood', 'neighbourhood_group', 'price']].groupby(['neighbourhood', 'neighbourhood_group']).mean()
```

```python
df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean()
```

- neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균을 계층적 indexing없이 구하고 NaN은 -999로 채우기

```python
df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean().unstack().fillna(-999)
```

- 데이터 중 neighbourhood_group 값이 Queens 값을 가지는 데이터 중 neighbourhood 그룹별로 price의 평균, 분산, 최대, 최소

```python
df[df.neighbourhood_group == 'Queens'].groupby('neighbourhood').price.agg(['mean', 'var', 'max', 'min']).head()
```

- neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율 구하기

```python
dfdf = df[['neighbourhood_group', 'room_type']].groupby(['neighbourhood_group', 'room_type']).size().unstack()
dfdf.head(4)
```

```python
dfdf.loc[:,:] = (dfdf.values / dfdf.sum(axis=1).values.reshape(-1, 1))
dfdf
```
