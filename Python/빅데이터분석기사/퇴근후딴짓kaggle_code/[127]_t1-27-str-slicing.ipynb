## basic1 데이터에서 f4가 E로 시작하면서 부산에 살고 20대인 사람은 몇 명일까요?

```python
import pandas as pd
df = pd.read_csv("../input/bigdatacertificationkr/basic1.csv")
df.head(1)
```

```python
df['E'] = df['f4'].str.startswith('E')
df.head()
```

```python
df[(df['city'] == '부산') & (df['E'] == True) & (df['age'] >= 20) & (df['age'] < 30)]
```

### 해설

```python
df['EI'] = df['f4'].str[:1] # 각 문자열에서 첫 번째 문자 추출
df.head(1)
```

```python
cond1 = df['EI'] == "E"
cond2 = df['city'] == "부산"
cond3 = (df['age'] >= 20) & (df['age'] < 30)

print(len(df[cond1 & cond2 & cond3]))
```
