## FILTERING & SORTING

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv')
df.head(1)
```

- quantity 컬럼 값이 3인 데이터를 추출하여 첫 5행 출력하기

```python
df.loc[df['quantity']==3].tail()
```

- quantity 컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하기

```python
df.loc[df['quantity'] == 3].head().reset_index(drop=True)
```

- quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하기

```python
df_1 = df[['quantity', 'item_price']]
df_1
```

- item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하기

```python
df['new_price'] = df_1['item_price'].str[1:].astype('float')
df['new_price'].head()
```

- new_price컬럼이 5 이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하기

```python
len(df.loc[df['new_price'] <= 5])
```

- item_name 명이 Chicken Salad Bowl인 데이터프레임을 추출하고 index 값을 초기화하기

```python
df.loc[df['item_name'] == 'Chicken Salad Bowl'].reset_index(drop=True)
```

- new_price값이 9이하이고 item_name값이 Chicken Salad Bowl인 데이터프레임을 추출하기

```python
df.loc[(df['new_price'] <= 9) & (df['item_name'] == 'Chicken Salad Bowl')]
```

- df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하기

```python
df.sort_values('new_price', ascending=False).reset_index(drop=True)
```

- df의 item_name 컬럼 값 중 Chips를 포함하는 경우의 데이터를 출력하기

```python
df.loc[df['item_name'].str.contains('Chips')]
```

- df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하기

```python
df.iloc[:, ::2]
```

- df의 item_name 컬럼값이 Steak Salad 또는 Bowl인 데이터를 인덱싱하기

```python
df.loc[(df.item_name == 'Steak Salad') | (df.item_name == 'Bowl')]
```

- df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 데이터 프레임화 한 후, item_name을 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남기기

```python
df_2 = df.loc[(df.item_name == 'Steak Salad') | (df.item_name == 'Bowl')]
df_2= df_2.drop_duplicates('item_name')
df_2
```

- df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 데이터 프레임화 한 후, item_name을 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남기기

```python
df_2 = df.loc[(df.item_name =='Steak Salad') | (df.item_name =='Bowl')]
df_2 = df_2.drop_duplicates('item_name',keep='last')
df_2
```

- df 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터를 인덱싱하기

```python
df.loc[df['new_price'] >= df['new_price'].mean()]
```

- df의 데이터 중 item_name의 값이 lzze인 데이터를 Fizzy Lizzy로 수정하기

```python
df.loc[df['item_name'] == 'Izze', 'item_name'] = 'Fizzy Lizzy'
df
```

- choice_description 값이 NaN인 데이터의 갯수

```python
df.choice_description.isnull().sum()
```

- choice_description 값이 NaN인 것을 NoData로 대체하기(loc이용)

- choice_description값에 Black이 들어가는 경우를 인덱싱하기

- choice_description 값에 Vegetables가 들어가지 않는 경우의 갯수
